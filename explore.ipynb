#### Imports

```python
import json
import plotly
import pandas as pd

from pymongo import MongoClient
import plotly.io as pio
from bunch import bunchify
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from collections import Counter
 
plotly.io.orca.config.executable = '/path/to/orca'
plotly.io.orca.config.save()
```

#### Connection to DB   

```python
client = MongoClient(port=27017)
db = client.FAIRsoft
alambique_coll = db.alambique
pretools_coll = db.pretools
```

#### Retrieval of tools

```python
def describe_collection(source):
    tools = []
    columns = set()
    for tool in alambique_coll.find({'@data_source':source}):
        [columns.add(col) for col in tool.keys()]
        tools.append(tool)

    tool_df = pd.DataFrame(tools, columns=columns)  
    tool_df.describe()
```

```python
describe_collection('bioconda')
```

```python
def describe_collection_pretools(source):
    tools = []
    columns = set()
    for tool in pretools_coll.find({'source':[source]}):
        [columns.add(col) for col in tool.keys()]
        tools.append(tool)

    tool_df = pd.DataFrame(tools, columns=columns)  
    tool_df.describe()
    return(tool_df.describe())
```

```python
describe_collection_pretools('bioconda_recipes')
```

```python
describe_collection_pretools('bioconda')
```
